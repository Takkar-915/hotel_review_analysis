全国の東横インさんの口コミデータを収集し、ネガポジ分析を行います。
これはデータ収集のためのスクリプトです。

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

file_path = "/content/drive/MyDrive/textfile/url_list.txt"

url_list = []

with open(file_path, "r", encoding="utf-8") as f:
  url_list = f.read().splitlines()  

print(url_list)

l=[]
l_2 = []
l_3 = []
point = 0

for url in url_list:
  r = requests.get(url)
  c = r.content

  soup = BeautifulSoup(c, "html.parser")

#口コミデータの収集
  reviews=soup.find_all("p",{"class":"jlnpc-kuchikomiCassette__postBody"})      
  

#星の数を収集
  stars = soup.find_all("div",{"class":"jlnpc-kuchikomiCassette__totalRate"})
  if not stars:
    break 

  

  for review in reviews:
      d={}
      d["クチコミ"]=review.text
      l.append(d)




  #ここで、ネガポジの指標として、星の数3以上をポジティブ、それ未満をネガティブなものとして捉え、ラベル付けを行う

  for star in stars:
      d_2 = 0
      d_2 = int(star.text)
      l_2.append(d_2)
      if d_2 > 3:
          point = 1
          l_3.append(point)
      else:
          point = 0
          l_3.append(point)    

#サーバへの負荷を考慮して3秒待機…
  time.sleep(3)

df = pd.DataFrame(l)

df["星"] = l_2

df["ネガポジ"] = l_3
  

df.to_csv("output.csv", encoding='utf_8_sig')

df
```
