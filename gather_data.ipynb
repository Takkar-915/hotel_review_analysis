じゃらんの旅館の口コミデータを収集するコード。今回は春休みに泊まった東横ＩＮＮ京都四条烏丸さんのデータを収集した。

```python
from numpy.ma.core import append
import requests
from bs4 import BeautifulSoup
import pandas as pd

r = requests.get("https://www.jalan.net/yad359432/kuchikomi/")
c = r.content

soup = BeautifulSoup(c, "html.parser")

all=soup.find_all("p",{"class":"jlnpc-kuchikomiCassette__postBody"})
all_2 = soup.find_all("div",{"class":"jlnpc-kuchikomiCassette__totalRate"})

l=[]
for item in all:
    d={}
    d["クチコミ"]=item.text
    l.append(d)

l_2 = []
l_3 = []
point = 0

for item_2 in all_2:
    d_2 = 0
    d_2 = int(item_2.text)
    l_2.append(d_2)
    if d_2 >= 3:
        point = 1
        l_3.append(point)
    else:
        point = 0
        l_3.append(point)    

df=pd.DataFrame(l)
#df.to_csv("review.csv", encoding="utf_8_sig")

df["星"] = l_2

df["ネガポジ"] = l_3

df.to_csv("output.csv", encoding='utf_8_sig')

df
```
